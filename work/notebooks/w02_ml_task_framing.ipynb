## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
import pandas as pd

# Pull data directly using your repository URL to avoid environment path issues
url = "https://raw.githubusercontent.com/kinzachoudhry8/my-ml-internship/main/data/raw/content_refresh_anonymized.csv"

try:
    df = pd.read_csv(url)
except Exception:
    # Fallback to local relative path if executing fully offline
    df = pd.read_csv("../data/raw/content_refresh_anonymized.csv")

# Display the grain of the data (The true Unit of Analysis)
print("--- UNIT OF ANALYSIS (GRAIN) ---")
print(f"Each row represents: A unique Content Asset (content_id) for a specific Client (client_id)\n")

# Isolate the core structural dataframe view required by the prompt
sample_view = df[['content_id', 'client_id', 'impressions_90d', 'sessions_90d', 'trend_direction']].head(5)
print(sample_view.to_string())

# Sketching the explicit target column vector representation
print("\n--- TARGET COLUMN SKETCH (is_declining_label) ---")
df['is_declining_label'] = df['trend_direction'].apply(lambda x: 1 if x == 'down' else 0)
print(df[['trend_direction', 'is_declining_label']].head(5).to_string())
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
# Check non-linear variance across clients to highlight why static rules fail
if 'client_id' in df.columns and 'impressions_90d' in df.columns:
    variance_check = df.groupby('client_id')['impressions_90d'].mean().head(3)
    print("Demonstrated Client Baseline Scale Variance:")
    print(variance_check)
```

**Why patterns are too messy for an if-statement:** A fixed engineering rule scales linearly and uniformly across all clients, which limits prioritization accuracy. For example, a hardcoded rule stating that any page with more than 180 days since its last update and over 500 impressions is a "refresh candidate" breaks down because different client profiles exhibit highly unbalanced traffic baselines and unique seasonal noise patterns.  Machine learning models (such as Random Forests or Gradient Boosted Trees) beat static rules by evaluating multi-dimensional correlations simultaneously—such as structural interactions between click-through rates, position tier decay, and user scroll engagement—ranking high-value opportunities with a massive lift in precision.
